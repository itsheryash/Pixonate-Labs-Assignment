```python
%matplotlib inline
pred_folder = r"C:\Users\lapto\Pixonate Project\runs\detect\predict"
label_folder = os.path.join(pred_folder, "labels")
image_paths = glob.glob(os.path.join(pred_folder, "*.jpg"))
for img_path in image_paths:
    base_name = os.path.splitext(os.path.basename(img_path))[0]
    label_path = os.path.join(label_folder, base_name + ".txt")
    if os.path.exists(label_path) and os.path.getsize(label_path) > 0:
        status = "Defective Image"
    else:
        status = "Normal Image"
    img = cv2.imread(img_path)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img_rgb)
    plt.title(status)
    plt.axis('off')
    plt.show()
```
